```
import Visionaire
from Visionaire.trainer import trainer
```

```
params = {
    'Image_Augmentation':'CIFAR10_A11_transformation',
    'data' : 'CIFAR10_dataloader',
    'model': 'Assign_11_model',
    'model_type':'EVA', # EVA or predefined
    'Batch_Size' : 512,
    'epochs':24,
    'criterion': 'NLLLoss',
    'L1_regularizer_lambda': 0,
    'L2_regularizer_lambda': 0.05,
    'optimizer': {
        'name':'SGD',
        'lr':0.001,
        'momentum':0.9
        #'weight_decay':0 L2_regularizer_lambda value will be automatically used here
                },
    'scheduler': {
         'name':'OneCycleLR',
         'max_at_epoch': 5
    }
}
```

```
my_trainer = trainer(params)
my_trainer.load_data()
```

```
#from torchsummary import summary

print(summary(my_trainer.model, input_size=(3, 32, 32)))
```

```
my_trainer.find_lr()
```

```
my_trainer.run(lr =0.01)
```

```
my_trainer.plot_lr()
```

```
my_trainer.plot_model_performance(False)
```

```
my_trainer.check_class_accuracy()
```

```
my_trainer.misclassified_images(False)
```
